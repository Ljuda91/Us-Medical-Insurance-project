# Initial Setup

```python
import csv
from statistics import mean, median

file = open("insurance.csv")
file_dict = csv.DictReader(file)
new_dict = dict()
total = 0

for row in file_dict:
    total += 1
    new_dict[total] = dict()
    new_dict[total].update(row)
file.close()

#Comma separator function
def cs(x):
    return f'{x:,.2f}'

#Upper function
def up(string):
    return string.upper()

clean_age = list()
clean_bmi = list()
clean_children = list()
clean_charges = list()

#Appending values from the dict to lists then converting them to proper variable types
for row in new_dict.values():
    clean_age.append(int(row['age']))
    clean_bmi.append(float(row['bmi']))
    clean_children.append(int(row['children']))
    clean_charges.append(float(row['charges']))

#Dictionary update with proper variable types
for i in range(len(clean_age)):
    new_dict[i+1]['age'] = clean_age[i]
    new_dict[i+1]['bmi'] = round(clean_bmi[i], 2)
    new_dict[i+1]['children'] = clean_children[i]
    new_dict[i+1]['charges'] = round(clean_charges[i], 2)
```

# Male vs Female Insurance total

```python
#Male vs Female total Insurance comparison
def male_vs_female(dictionary):
    
    female_charges_total = list()
    
    for i in dictionary.values():
        if i['sex'] == 'female':
            female_charges_total.append(i['charges'])
            
    result_f = f'''The total insurance charges for all {len(female_charges_total)} {up('female')} insurees is: {(cs(sum(female_charges_total)))}$.
The median charges are: {cs(median(female_charges_total))}$.'''

    male_charges_total = list()
    
    for i in dictionary.values():
        if i['sex'] == 'male':
            male_charges_total.append(i['charges'])
            
    result_m = f'''
The total insurance charges for all {len(male_charges_total)} {up('male')} insurees is: {(cs(sum(male_charges_total)))}$.
The median charges are: {cs(median(male_charges_total))}$.'''
    return f'''
{result_f}
{result_m}
'''
male_vs_female_test = male_vs_female(new_dict)
print(male_vs_female_test)
```

# Smoker vs Non-Smoker Insurance total

```python
#Smokers and non-smokers Insurance comparison
def smoker_vs_non(dictionary):
    smoker_total = list()
    non_smoker_total = list()
    for i in dictionary.values():
        if i['smoker'] == 'yes':
            smoker_total.append(i['charges'])
            
        elif i['smoker'] == 'no':
            non_smoker_total.append(i['charges'])
            
    result = f'''
There is a total of {len(smoker_total)} {up('smokers')} and their total charges are: {cs(sum(smoker_total))}$.
The median charges are: {cs(median(smoker_total))}$.

There is a total of {len(non_smoker_total)} {up('non-smokers')} and their total charges are: {cs(sum(non_smoker_total))}$.
The median charges are: {cs(median(non_smoker_total))}$.
'''
    return result
smokers_comp = smoker_vs_non(new_dict)
print(smokers_comp)
```

# Youngest with the most expensive insurance and least expensive insurance

```python
#Finding the youngest Insurees with max and min insurance charges
def youngest():
    youngest = 18
    total_list = list()
    id_min = None
    id_max = None
    for i in new_dict.values():
        if i['age'] < youngest:
            youngest = i['age']
    for i in new_dict.values():        
        if i['age'] == youngest:
            total_list.append(i['charges'])
    
    for i in new_dict.values():
        if i['age'] == youngest and i['charges'] == min(total_list):
            id_min = i
        elif i['age'] == youngest and i['charges'] == max(total_list):
            id_max = i
    result = f'The youngest person with the highest charges has these results: \n{id_max}, \nand the youngest with the lowest charges: \n{id_min}'
    return result

young_test = youngest()
print(young_test)
```

# Oldest with the most and least expensive insurance

```python
#Finding the oldest Insurees with max and min insurance charges
def oldest():
    oldest = 18
    total_list = list()
    id_min = None
    id_max = None
    for i in new_dict.values():
        if i['age'] > oldest:
            oldest = i['age']
    for i in new_dict.values():    
        if i['age'] == oldest:
            total_list.append(i['charges'])
    
    for i in new_dict.values():
        if i['age'] == oldest and min(total_list) == i["charges"]:
            id_min = i
        elif i['age'] == oldest and max(total_list) == i['charges']:
            id_max = i
    result = f'The oldest person with the highest charges has these results: \n{id_max}, \nand the oldest with the lowest charges: \n{id_min}'
    return result
    
oldest_test = oldest()
print(oldest_test)
```

# Children influence on Insurance

```python
#Comparing non-smokers with children and without
def children_no_smokers(dictionary):
    total_chil = list()
    
    total_no_chil = list()
    
    for i in dictionary.values():
        if i['smoker'] == 'no' and i['children'] > 0:
            total_chil.append(i['charges'])
            
        elif i['smoker'] == 'no' and i['children'] == 0:
            total_no_chil.append(i['charges'])
            
    result = f'''
There are a total of {len(total_chil)} {up('non-smoking')} people with {up('children')} and their total is: {cs(sum(total_chil))}$.
The median charges are: {cs(median(total_chil))}$.

There are a total of {len(total_no_chil)} {up('non-smoking')} people with {up('no children')} and their total is: {cs(sum(total_no_chil))}$.
The median charges are: {cs(median(total_no_chil))}$.'''
    return result
children = children_no_smokers(new_dict)
print(children)

#Same thing just with smokers
def children_yes_smokers(dictionary):
    total_chil = list()
    total_no_chil = list()
    for i in dictionary.values():
        if i['smoker'] == 'yes' and i['children'] > 0:
            total_chil.append(i['charges'])
        elif i['smoker'] == 'yes' and i['children'] == 0:
            total_no_chil.append(i['charges'])
            
    result = f'''
There are a total of {len(total_chil)} {up('smoking')} people with {up('children')} and their total is: {cs(sum(total_chil))}$.
The median charges are: {cs(median(total_chil))}$.

There are a total of {len(total_no_chil)} {up('smoking')} people with {up('no children')} and their total is: {cs(sum(total_no_chil))}$.
The median charges are: {cs(median(total_no_chil))}$.
'''
    return result
children_smokers =children_yes_smokers(new_dict)
print(children_smokers)
```

# Ultimate checklist for Insurance

```python
#Search for a very nieche group of people and results of parallel results
def ultimate(age, sex, children, smoker, region):
    total1 = list()
    total2 = list()
    sex_assign = ''
    smoker_assign = ''
    for i in new_dict.values():
        if age == i['age'] and sex == i['sex'] and smoker == i['smoker'] and children == i['children'] and region == i['region']:
            total1.append(i['charges'])
    
        elif age == i['age'] and i['sex'] != sex and i['smoker'] != smoker and children == i['children'] and region != i['region']:
            total2.append(i['charges'])
            sex_assign = i['sex']
            smoker_assign = i['smoker']

    result = f'''
The results for this group are:
    - This group of people that are {age} years old, are a {up(sex)} and have {children} child/ren. 
    - They smoke: {up(smoker)} and are from this {up(region)} region. 
    - Their total is: {cs(sum(total1))}$. 
    - There is a total of {len(total1)} people.
    
Results for the opposite {up('sex')} group are:
    - This group of people that are {age} years old, are a {up(sex_assign)} and have {children} child/ren. 
    - They smoke: {up(smoker_assign)} and are from all other regions. 
    - Their total is: {cs(sum(total2))}$. 
    - There is a total of {len(total2)} people. 
    '''
    return result
group = ultimate(35, 'female', 1, 'no', 'northwest')
print(group)
```

## Age range function

```python
#Looking at the results for a single age group or a range between two ages 
age_range = []

for i in new_dict.values():
        if i['age'] not in age_range:
            age_range.append(i['age'])
sorted_age = sorted(age_range)

def age_range_function(min_age, max_age):
    total_ns_nc = list()
    
    total_ns_yc = list()
    
    total_ys_nc = list()
    
    total_ys_yc = list()
    
    for i in new_dict.values():
        if min_age not in age_range or max_age not in age_range:
            print('The age ranges you entered are not valid. Please try again.')
            break
        elif min_age > max_age:
            print('MIN age is higher than MAX age. Please try again')
            break
        elif i['age'] >= min_age and i['age'] <= max_age and i['smoker'] == 'no' and i['children'] == 0:
            total_ns_nc.append(i['charges'])

        elif i['age'] >= min_age and i['age'] <= max_age and i['smoker'] == 'no' and i['children'] > 0:
            total_ns_yc.append(i['charges'])
            
        elif i['age'] >= min_age and i['age'] <= max_age and i['smoker'] == 'yes' and i['children'] == 0:
            total_ys_nc.append(i['charges'])
            
        elif i['age'] >= min_age and i['age'] <= max_age and i['smoker'] == 'yes' and i['children'] > 0:
            total_ys_yc.append(i['charges'])
            
    result = f'''
The results for the age range from {min_age} to {max_age} are: 

- There are {len(total_ns_nc) + len(total_ns_yc) + len(total_ys_nc) + len(total_ys_yc)} people within this age range.
- The total charges are: {cs(sum(total_ns_nc + total_ns_yc + total_ys_nc + total_ys_yc))}$.

    - There is {len(total_ns_nc)} {up('non-smokers')} with {up('no children')} within this age range.
    - The groups total charges are: {cs(sum(total_ns_nc))}$. 
    - The groups median is: {cs(median(total_ns_nc))}$. 
    
    - There is {len(total_ns_yc)} {up('non-smokers')} with {up('children')}.
    - The groups total charges are: {cs(sum(total_ns_yc))}$. 
    - The groups median is: {cs(median(total_ns_yc))}$. 

    - There is {len(total_ys_nc)} {up('smokers')} with {up('no children')}.
    - The groups total charges are: {cs(sum(total_ys_nc))}$. 
    - The groups median is: {cs(median(total_ys_nc))}$. 

    - There is {len(total_ys_yc)} {up('smokers')} with {up('children')}.
    - The groups total charges are: {cs(sum(total_ys_yc))}$. 
    - The groups median is: {cs(median(total_ys_yc))}$.
    '''
    return result

age18_24 = age_range_function(18, 24)
age25_34 = age_range_function(25, 34)
age35_44 = age_range_function(35, 44)
age45_54 = age_range_function(45, 54)
age55_64 = age_range_function(55, 64)

print(age18_24, age35_44)
```

# Region function

```python
def region_calc(region):
    total_ns_nc = list()
    
    total_ns_yc = list()
    
    total_ys_nc = list()
    
    total_ys_yc = list()

    for i in new_dict.values():
        if i['smoker'] == 'no' and i['children'] == 0 and i['region'] == region:
            total_ns_nc.append(i['charges'])
            
        elif i['smoker'] == 'no' and i['children'] > 0 and i['region'] == region:
            total_ns_yc.append(i['charges'])
            
        elif i['smoker'] == 'yes' and i['children'] == 0 and i['region'] == region:
            total_ys_nc.append(i['charges'])
            
        elif i['smoker'] == 'yes' and i['children'] > 0 and i['region'] == region:
            total_ys_yc.append(i['charges'])
            
    result = f'''
The results for the {up(region)} region are: 

- There are: {len(total_ns_nc) + len(total_ns_yc) + len(total_ys_nc) + len(total_ys_yc)} people in that region.
- The total charges are: {cs(sum(total_ns_nc) + sum(total_ns_yc) + sum(total_ys_nc) + sum(total_ys_yc))}$.
- The {up(region)}'s region average charges are: {cs(mean(total_ns_nc + total_ns_yc + total_ys_nc + total_ys_yc))}$ and median: {cs(median(total_ns_nc + total_ns_yc + total_ys_nc + total_ys_yc))}$.
    
    - There is {len(total_ns_nc)} {up('non-smokers')} with {up('no children')}.
    - The total for this group is: {cs(sum(total_ns_nc))}$. 
    - The groups median is: {cs(median(total_ns_nc))}$. 
    
    - There is {len(total_ns_yc)} {up('non-smokers')} with {up('children')}.
    - The total for this group is: {cs(sum(total_ns_yc))}$. 
    - The groups median is: {cs(median(total_ns_yc))}$. 

    - There is {len(total_ys_nc)} {up('smokers')} with {up('no children')}.
    - The total for this group is: {cs(sum(total_ys_nc))}$. 
    - The groups median is: {cs(median(total_ys_nc))}$. 
    
    - There is {len(total_ys_yc)} {up('smokers')} with {up('children')}.
    - The total for this group is: {cs(sum(total_ys_yc))}$. 
    - The groups median is: {cs(median(total_ys_yc))}$. 
'''
    return result

def regions_total(dictionary):
    reg_total1 = list()
    reg_total2 = list()
    reg_total3 = list()
    reg_total4 = list()

    for i in dictionary.values():
        if i['region'] == 'southwest':
            reg_total1.append(i['charges'])
            
        elif i['region'] == 'southeast':
            reg_total2.append(i['charges'])
            
        elif i['region'] == 'northwest':
            reg_total3.append(i['charges'])
            
        elif i['region'] == 'northeast':
            reg_total4.append(i['charges'])
            
    result = f'''
The {up('southwest')} region has {len(reg_total1)} and total charges: {cs(sum(reg_total1))}$.
The median charges are: {cs(median(reg_total1))}$.

The {up('southeast')} region has {len(reg_total2)} and total charges: {cs(sum(reg_total2))}$.
The median charges are: {cs(median(reg_total2))}$.

The {up('northwest')} region has {len(reg_total3)} and total charges: {cs(sum(reg_total3))}$.
The median charges are: {cs(median(reg_total3))}$.

The {up('northeast')} region has {len(reg_total4)} and total charges: {cs(sum(reg_total4))}$.
The median charges are: {cs(median(reg_total4))}$.
    '''
    return result
southeast = region_calc('southeast')
southwest = region_calc('southwest')
northeast = region_calc('northeast')
northwest = region_calc('northwest')

regions = regions_total(new_dict)

print(regions, southeast)
```
